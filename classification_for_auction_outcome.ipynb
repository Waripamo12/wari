```python
from scipy.stats import zscore
```

```python
# Detecting and handling outliers using Z-score
z_scores = np.abs(zscore(df[numeric_cols]))
df = df[(z_scores < 3).all(axis=1)]
```

```python
# Normalize (Example: Min-Max Scaling)
# Adjust as per need
for col in numeric_cols:
    df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
```

```python
# Feature Interaction (Example: b1*b2)
df['b1_b2_interaction'] = df['process.b1.capacity'] * df['process.b2.capacity']
```

```python
df['verification.result'].value_counts(normalize = True)  #imbalance class on the result with higher 87% is False, 12% of True
```

```python
#converting the category features to numeric
df['verification.result'] =df['verification.result'].astype('category').cat.codes
```

```python
#splitting the dataset
X = df.drop('verification.result',axis=1)
y =df['verification.result']
```

```python
#using the SMOTE to handle the imbalance class of the dataset
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
x_smote, y_smote = smote.fit_resample(X, y)

print('Original dataset shape', y.shape)
print('Resample dataset shape', y_smote.shape)
```

```python
#importing and normalizing the numeric features toa range of 0 and 1
from sklearn.preprocessing import normalize

X = normalize(X)
x_smote = normalize(x_smote)
```

```python
from sklearn.model_selection import train_test_split
# i.e. 70 % training dataset and 30 % test datasets
X_train, X_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size = 0.30)
```

```python
X_train.shape
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_predict, train_test_split, KFold
from sklearn.metrics import roc_curve, classification_report, confusion_matrix, auc
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# 3. Building Models
rf = RandomForestClassifier(n_estimators=100, random_state=42)
lr = LogisticRegression(random_state=42)

# 4. Cross-Validation with KFold
cv = KFold(n_splits=5, shuffle=True, random_state=42)
rf_probs = cross_val_predict(rf, X_train, y_train, cv=cv, method="predict_proba")[:, 1]
lr_probs = cross_val_predict(lr, X_train, y_train, cv=cv, method="predict_proba")[:, 1]

# 5. ROC Curve and Save Image
fpr_rf, tpr_rf, _ = roc_curve(y_train, rf_probs)
fpr_lr, tpr_lr, _ = roc_curve(y_train, lr_probs)

plt.figure(figsize=(10, 6))
plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC = {auc(fpr_rf, tpr_rf):.4f})")
plt.plot(fpr_lr, tpr_lr, label=f"Logistic Regression (AUC = {auc(fpr_lr, tpr_lr):.4f})")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.savefig("ROC_Curve.png")
plt.show()

# 6. Train on Full Data and Evaluate on Test Set
rf.fit(X_train, y_train)
lr.fit(X_train, y_train)

rf_preds = rf.predict(X_test)
lr_preds = lr.predict(X_test)

# 7. Print Classification Reports
print("Random Forest Classification Report:\n", classification_report(y_test, rf_preds))
print("\nLogistic Regression Classification Report:\n", classification_report(y_test, lr_preds))

# 8. Plot and Save Confusion Matrices
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(confusion_matrix(y_test, rf_preds), annot=True, fmt='d', ax=ax[0], cmap='Blues')
ax[0].set_title("Random Forest Confusion Matrix")
ax[0].set_xlabel("Predicted label")
ax[0].set_ylabel("True label")

sns.heatmap(confusion_matrix(y_test, lr_preds), annot=True, fmt='d', ax=ax[1], cmap='Blues')
ax[1].set_title("Logistic Regression Confusion Matrix")
ax[1].set_xlabel("Predicted label")
ax[1].set_ylabel("True label")

plt.tight_layout()
plt.savefig("Confusion_Matrices.png")
plt.show()
```
